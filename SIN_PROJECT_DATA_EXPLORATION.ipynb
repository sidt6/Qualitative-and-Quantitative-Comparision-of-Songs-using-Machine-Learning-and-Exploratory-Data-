```python
plt.scatter(data_frame['key'], data_frame['tempo'], c=data_frame['time_signature'], 

            s=data_frame['tempo'])
 
# Adding Title to the Plot

plt.title("Scatter Plot")
 
# Setting the X and Y labels

plt.xlabel('Key')

plt.ylabel('Tempo')
 
plt.colorbar()
 
plt.show()
```

```python
sns.histplot(x='popularity', data=data_frame, kde=True, hue='like')
 
plt.show()
```

```python
sns.lineplot(x='popularity', y='like', data=data_frame)
plt.show()
```

```python
sns.barplot(x='key',y='popularity', data=data_frame, 

            hue='like')
 
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(data_frame['danceability'])
plt.title('Danceability of songs', fontsize=18)
plt.show()
```

```python
features = ['energy', 'key', 'mode', 'tempo']
plt.subplots(figsize=(30,15))
import seaborn as sb 
for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  sb.distplot(data_frame[col])
plt.show()
```

```python
#Using Pearson Correlation
plt.figure(figsize=(20,10))
cor = data_frame.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```
